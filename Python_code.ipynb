# Chi-Square Test

```python
cat_cols=['Nationality', 'State', 'Region',
       'Social_Media_presence', 'International_Travel_History', 'Gender',
       'Job', 'Marital Status',
       'Education', 'Default', 'Housing', 'Loan','Contact',
         'poutcome']
y='target_variable'
```

```python
import scipy
for i in cat_cols:
    tmp=pd.crosstab(df[i],df["target_variable"],margins=True)
    lis=[]
    for j in range(tmp.shape[0]-1):
        x=tmp.shape[1]-1
        lis.append(list(tmp.iloc[j][0:x].values))
#         print(lis)
    obs=np.array(lis)
#     print(obs)
    print("for",i)
    print("CHi-Square Statistic",scipy.stats.chi2_contingency(obs)[0])
    print("p-value",scipy.stats.chi2_contingency(obs)[1])
    print("Degrees of Freedom",scipy.stats.chi2_contingency(obs)[2],"\n","-"*50)
    
    print("-"*100,"\n\n")
```

# Removing Outliers

```python
lower_boundries= []
upper_boundries= []
for i in ["Num_dependants","Months_at_Current_address","Age", "Duration", "contacts_during_campaign"]:
    IQR= df[i].quantile(0.75) - df[i].quantile(0.25)
    lower_band= df[i].quantile(0.25) - (1.5*IQR)
    upper_band= df[i].quantile(0.75) + (1.5*IQR)
    
    print(i, ":", lower_band, ",",  upper_band)
    
    lower_boundries.append(lower_band)
    upper_boundries.append(upper_band)
```

```python
j = 0
for i in ["Num_dependants","Months_at_Current_address","Age", "Duration", "contacts_during_campaign"]:
    df.loc[df[i] > upper_boundries[j], i] = int(upper_boundries[j])
    j = j + 1  
```

# Label Encoding for Machine Learning Algorithms 

```python
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()
cat_cols=['Social_Media_presence', 'International_Travel_History', 'Gender',
       'Job', 'Marital Status',"Month",
       'Education', 'Default', 'Housing', 'Loan', 'Contact',
       'poutcome']

for st in cat_cols:
    df[st]=labelencoder.fit_transform(df[st])
```

# Dividing into X & Y Variables

```python
x=df.drop(["target_variable"],axis=1)
y=df["target_variable"]
```

# Oversampling for imbalanced Dataset

```python
from imblearn.combine import SMOTETomek

smote= SMOTETomek(0.75)

x_sm, y_sm= smote.fit_resample(x,y)

# x_sm and y_sm are the resampled data

# target class count of resampled dataset
y_sm.value_counts()
```

# Importing Modelling Libraries

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x_sm, y_sm, test_size=0.3)
```

# Logistic Regression

```python
logistic_model=LogisticRegression()

logistic_param={"C": 10.0**np.arange(-2,3), "penalty": ["l1", "l2"]}

cv_log= StratifiedKFold(n_splits=10)

gridsearch_logistic= GridSearchCV(logistic_model, logistic_param, cv=cv_log, scoring= "f1_macro", n_jobs=-1, verbose=2)
gridsearch_logistic.fit(x_train, y_train)

print("best score is:" ,gridsearch_logistic.best_score_)
print("best parameters are:" ,gridsearch_logistic.best_params_)
```

```python
logistic_predicted= gridsearch_logistic.predict(x_test)

cm= confusion_matrix(y_test, logistic_predicted)
print(cm)
print(accuracy_score(y_test, logistic_predicted))
print(classification_report(y_test, logistic_predicted))
```

# K-Nearest Neighbour 

```python
knn_model=KNeighborsClassifier()

knn_model.fit(x_train, y_train)
```

```python
knn_predicted= knn_model.predict(x_test)

cm= confusion_matrix(y_test, knn_predicted)
print(cm)
print(accuracy_score(y_test, knn_predicted))
print(classification_report(y_test, knn_predicted))
```

```python
from sklearn.model_selection import cross_val_score
print("KNN",np.average(cross_val_score(KNeighborsClassifier(),x_train,y_train,cv=10)))
```

# Decision Tree Model

```python
dt_model= DecisionTreeClassifier()

dt_param= { 
           "max_features": ["auto", "sqrt", "log2"],
           "max_depth": [int(x) for x in np.linspace(start=5, stop=30, num=6)],
           "min_samples_split": [5,10,15,100],
           "min_samples_leaf": [1,2,5,10],
           "criterion":['gini', 'entropy'] 
          }

cv_dt= StratifiedKFold(n_splits=5)

randomsearch_dt= RandomizedSearchCV(dt_model, dt_param, cv=cv_dt, scoring= "f1_macro", n_jobs=-1, verbose=2, n_iter=10)

randomsearch_dt.fit(x_train, y_train)

print("best score is:", randomsearch_dt.best_score_)
print("best parameters are:", randomsearch_dt.best_params_)

```

```python
dt_predicted= randomsearch_dt.predict(x_test)

cm= confusion_matrix(y_test, dt_predicted)
print(cm)
print(accuracy_score(y_test, dt_predicted))
print(classification_report(y_test, dt_predicted))
```

# Random Forest Model

```python
rf_model= RandomForestClassifier()

rf_param= { 
           "n_estimators": [int(x) for x in np.linspace(start=100, stop=1000, num=10)],
           "max_features": ["auto", "sqrt", "log2"],
           "max_depth": [int(x) for x in np.linspace(start=5, stop=30, num=6)],
           "min_samples_split": [5,10,15,100],
           "min_samples_leaf": [1,2,5,10],
           "criterion":['gini', 'entropy'] 
          }

cv_rf= StratifiedKFold(n_splits=5)

randomsearch_rf= RandomizedSearchCV(rf_model, rf_param, cv=cv_rf, scoring= "f1_macro", n_jobs=-1, verbose=2, n_iter=10)

randomsearch_rf.fit(x_train, y_train)

print("best score is:", randomsearch_rf.best_score_)
print("best parameters are:", randomsearch_rf.best_params_)

```

```python
rf_predicted= randomsearch_rf.predict(x_test)

cm= confusion_matrix(y_test, rf_predicted)
print(cm)
print(accuracy_score(y_test, rf_predicted))
print(classification_report(y_test, rf_predicted))
```

Since, Accuracy, Precision and F1 Score of Random forest Model is better than other models. We have selected random forest model
